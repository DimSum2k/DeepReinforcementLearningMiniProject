# Context

In a reinforcement learning algorithm, we modelize each step $t$ as an action $a_t$ obtained from a state $s_t$, i.e. $\{(a_{t},s_{t})_{t\leq T}\}$ having the Markov property. We consider a discount factor $\gamma \in [0,1]$ that ensures convergence. The goal is to find among all the policies $\pi$, $\pi(a|s)$, one that maximizes the expected reward:

\begin{equation*}
R(\pi)=\sum_{t\leq T}E_{p^{\pi}}[\gamma^t r(s_{t},a_{t})] \> ,
\end{equation*}

where: 
\begin{equation*}p^{\pi}(a_{0},a_{1},s_{1},...,a_{T},s_{T})=p(a_{0})\prod_{t=1}^{T}\pi(a_{t}|s_{t})p(s_{t+1}|s_{t},a_{t}) \> .
\end{equation*}

We note the $Q$-function:

\begin{equation*}Q^\pi(s,a)=E_{p^{\pi}}[\sum_{t\leq T}\gamma^{t}r(s_{t},a_{t})|s_{0}=s,a_{0}=a] \> .
\end{equation*}

Thus, the optimal Q function is:
\begin{equation*}
Q^*(s,a)=\max_{\pi}Q^\pi(s,a) \> .
\end{equation*}

In this project, we will apply the deep reinforcement learning techniques to a simple game: an agent will have to learn from scratch a policy that will permit it maximizing a reward.

### The environment

```Environment``` is an abstract class that represents the states, rewards, and actions to obtain the new state.

````python
class Environment(object):
    def __init__(self):
        pass

    def act(self, act):
        """
        One can act on the environment and obtain its reaction:
        - the new state
        - the reward of the new state
        - should we continue the game?

        :return: state, reward, game_over
        """
        pass


    def reset(self):
        """
        Reinitialize the environment to a random state and returns
        the original state

        :return: state
        """
        pass
    
    def draw(self):
        """
        Visualize in the console or graphically the current state
        """
        pass
    
````

The method ```act``` allows to act on the environment at a given state $s_t$ (stored internally), via action $a_t$. The method will return the new state $s_{t+1}$, the reward $r(s_{t},a_{t})$ and determines if $t\leq T$ (*game_over*).

The method ```reset``` simply reinitializes the environment to a random state $s_0$.

The method ```draw``` displays the current state $s_t$ (this is useful to check the behavior of the Agent).

We modelize $s_t$ as a tensor, while $a_t$ is an integer.

__Question 1__:
Explain the function act. Why is ```epsilon``` essential?

The function `act` implements an $\epsilon$-greedy policy. With probability $\epsilon$ the agent takes an action uniformely at random and with probability $1-\epsilon$ the agent takes an action according to its internal current policy $\pi$. 

An $\epsilon$-greedy policy with $\epsilon>0$ is a good way to maintain a good exploration-exploitation trade off. If $\epsilon$ is close to 1 the agent will most of the time explores the environment by taking random actions while if $\epsilon$ is close to 0 the agent will most of the time exploits what it already knows by following its learned policy. A good strategy is thus to start the training with a large $\epsilon$ and to impose a decay on $\epsilon$ throughout the training. It is sometimes a good idea to maintain $\epsilon$ at a positive value at the end of the decay because it is most of the time a good idea to maintain a bit of exploration even towards the end of the training. 

When we evaluate the agent we should not forget to set $\epsilon$ to zero, that is why the function `act` has a `train` argument. However, in practice, we see that sometimes letting a small epsilon value even at test time is beneficial.

# The game, *eat cheese*

A rat runs on an island and tries to eat as much as possible. The island is subdivided into $N\times N$ cells, in which there are cheese (+0.5) and poisonous cells (-1). The rat has a visibility of 2 cells (thus it can see $5^2$ cells). The rat is given a time $T$ to accumulate as much food as possible. It can perform 4 actions: going up, down, left, right. 

The goal is to code an agent to solve this task that will learn by trial and error. We propose the following environment:

````python
class Environment(object):
    def __init__(self, grid_size=10, max_time=500, temperature=0.1):
        
        ## size of the map (including limits)
        grid_size = grid_size+4
        self.grid_size = grid_size
        ## max number of actions
        self.max_time = max_time
        ## probability of having a cheesy cell or a poisonous cell
        self.temperature = temperature

        #board on which one plays  
        ## store items
        self.board = np.zeros((grid_size,grid_size))
        ## store positions (rat pos. and game limits)
        self.position = np.zeros((grid_size,grid_size))

        # coordinate of the rat
        self.x = 0
        self.y = 1

        # self time
        self.t = 0

        ## Parameter for the display
        self.scale=16

        ## store every step of the game for video displaying
        self.to_draw = np.zeros((max_time+2, grid_size*self.scale, grid_size*self.scale, 3))

    def draw(self,e):
        """saves the game as a mp4 video"""
        skvideo.io.vwrite(str(e) + '.mp4', self.to_draw)
        
    def get_frame(self,t):
        '''Called before every taken action to store the game as an array'''
        
        b = np.zeros((self.grid_size,self.grid_size,3))+128
        b[self.board>0,0] = 256        ## channel0=256 -> cheese = RED
        b[self.board < 0, 2] = 256     ## channel2=256 -> poisonous = BLUE
        b[self.x,self.y,:]=256         ## channel1/2/3=256 -> rat position = WHITE
        b[-2:,:,:]=0                   ## 1/2/3channel = 0 -> limit of the map = BLACK
        b[:,-2:,:]=0
        b[:2,:,:]=0
        b[:,:2,:]=0
        b =  cv2.resize(b, None, fx=self.scale, fy=self.scale, interpolation=cv2.INTER_NEAREST)

        self.to_draw[t,:,:,:]= b ## save frame t to the array


    def act(self, action):
        """This function returns the new state, reward and decides if the
        game ends.
        Actions:
                        N.1
                    O.3     E.2
                        S.0
        """

        ## Update video array
        self.get_frame(int(self.t)) 

        self.position = np.zeros((self.grid_size, self.grid_size))

        ## Set the limits  of the game (-1)
        self.position[0:2,:]= -1
        self.position[:,0:2] = -1
        self.position[-2:, :] = -1
        
        # ADD
        #self.position[-2:, :] = -1 
        self.position[:, -2:] = -1
        # ADD

        ## Rat position is 1
        self.position[self.x, self.y] = 1 
        if action == 0: ## Go S except if you touch the limit then go N (repulsive move)
            if self.x == self.grid_size-3:
                self.x = self.x-1
            else:
                self.x = self.x + 1
        elif action == 1: ## Go N except if you touch the limit then go S (repulsive move)
            if self.x == 2:
                self.x = self.x+1
            else:
                self.x = self.x-1
        elif action == 2: ## Go E except if you touch the limit then go O (repulsive move)
            if self.y == self.grid_size - 3:
                self.y = self.y - 1
            else:
                self.y = self.y + 1
        elif action == 3: ## Go O except if you touch the limit then go E (repulsive move)
            if self.y == 2:
                self.y = self.y + 1
            else:
                self.y = self.y - 1
        else:
            RuntimeError('Error: action not recognized')

        self.t = self.t + 1               
        reward = self.board[self.x, self.y] ## get the reward on the new position
        self.board[self.x, self.y] = 0      ## remove the item from the map 
        game_over = self.t > self.max_time  
        state = np.concatenate((self.board.reshape(self.grid_size, self.grid_size,1),
                        self.position.reshape(self.grid_size, self.grid_size,1)),axis=2)
        state = state[self.x-2:self.x+3,self.y-2:self.y+3,:] # visibility zone of the rat is 5*5

        return state, reward, game_over

    def reset(self):
        """This function resets the game and returns the initial state"""

        ## random x-axis and y-axis positions for the rat (excluding limit zone)
        self.x = np.random.randint(3, self.grid_size-3, size=1)[0]
        self.y = np.random.randint(3, self.grid_size-3, size=1)[0]

        ## set the bonus (cheese +0.5) on the map (probability = temperature)
        bonus = 0.5*np.random.binomial(1,self.temperature,size=self.grid_size**2)
        bonus = bonus.reshape(self.grid_size,self.grid_size)

        ## set the malus (poisonous cells -1.) on the map (probability = temperature)
        malus = -1.0*np.random.binomial(1,self.temperature,size=self.grid_size**2)
        malus = malus.reshape(self.grid_size, self.grid_size)

        ## array to fill for video display
        self.to_draw = np.zeros((self.max_time+2, self.grid_size*self.scale, self.grid_size*self.scale, 3))

        ## prevent cells from being both poisonous and have a cheese (cheese >> poisonous)
        malus[bonus>0]=0
    
        self.board = bonus + malus

        ## store game limits and rat position
        self.position = np.zeros((self.grid_size, self.grid_size))
        self.position[0:2,:]= -1 
        self.position[:,0:2] = -1
        self.position[-2:, :] = -1
        
        # ADD
        # self.position[-2:, :] = -1
        self.position[:, -2:] = -1 
        # ADD
        
        ## prevent the rat from starting on a poisonous/cheesy cell
        self.board[self.x,self.y] = 0 
        self.t = 0
        
        ## ADD - position of the rat at the beginning
        self.position[self.x,self.y] = 1
        ## ADD

        # state is the concatenation of both the board (cheese/poisonous) and the positions
        state = np.concatenate((
                               self.board.reshape(self.grid_size, self.grid_size,1),
                        self.position.reshape(self.grid_size, self.grid_size,1)),axis=2)
        
        # the rat only sees a 5*5 window around its position
        state = state[self.x - 2:self.x + 3, self.y - 2:self.y + 3, :]
                
        return state
    
````

````python
def test(agent, env, epochs, prefix='', plot=False, save=True):
    
    # Number of won games
    score = 0
    
    score_per_epoch = [] 
        
    for e in range(1, epochs + 1):
        # At each epoch, we restart to a fresh game and get the initial state
        state = env.reset()
        # This assumes that the games will end
        game_over = False

        win = 0
        lose = 0

        while not game_over:
            # The agent performs an action
            action = agent.act(state, train=False)

            # Apply an action to the environment, get the next state, the reward
            # and if the game end
            prev_state = state
            state, reward, game_over = env.act(action)

            # Update the counters
            if reward > 0:
                win = win + reward
            if reward < 0:
                lose = lose -reward
        
        # Save as a mp4
        if save:
            env.draw(prefix + str(e))

        # Update stats
        score += win - lose
        score_per_epoch.append(win-lose)

        print("{}, Win/lose count {}/{}. Average score ({})".format(e,win, lose, score / e))
    final_score = score / epochs
    print('Final score: {}'.format(final_score))

    if plot:
        plt.figure()
        plt.plot(score_per_epoch)
        plt.title("Score per Epoch")
        plt.xlabel("epoch")
        plt.ylabel("score")
        
    return score_per_epoch, final_score
````

\begin{equation*}

The correct statement should be, 
<font color=red> \begin{equation*}
Q^{\pi}(s,a)=E_{\left(s^{\prime}, a^{\prime}\right) \sim p^{\pi}(\cdot,\cdot | s, a)}[r(s,a)+\gamma Q^{\pi}(s',a')]
\end{equation*}
    where $p^{\pi}(s^{\prime}, a^{\prime} | s, a) = \pi(a^{\prime} | s^{\prime})p(s^{\prime} | s, a)$
</font>

If $T=\infty$, we have, \begin{equation*}
Q^\pi(s,a)=E_{p^{\pi}}[\sum_{t\geq 0}\gamma^{t}r(s_{t},a_{t})|s_{0}=s,a_{0}=a]
\end{equation*}

Thus, 
\begin{equation*}
\begin{aligned} 
Q^{\pi}(s, a) &=\mathbb{E}_{p^{\pi}}\left[\sum_{t = 0}^{\infty} \gamma^{t} r\left(s_{t}, a_{t}\right) | s_{0}=s, a_{0}=a\right] \\ &=\mathbb{E}_{p^{\pi}}\left[r(s, a)+\sum_{t = 1}^{\infty} \gamma^{t} r\left(s_{t}, a_{t}\right) | s_{0}=s, a_{0}=a\right] \\ &=r(s, a)+\gamma \mathbb{E}_{p^{\pi}}\left[\sum_{t =1}^{\infty} \gamma^{t-1} r\left(s_{t}, a_{t}\right) | s_{0}=s, a_{0}=a\right] \\ 
\end{aligned}
\end{equation*}

We develop the second term,

\begin{equation*}
\begin{aligned} 
& \mathbb{E}_{p^{\pi}}\left[\sum_{t =1}^{\infty} \gamma^{t-1} r\left(s_{t}, a_{t}\right) | s_{0}=s, a_{0}=a\right] \\ &=\sum_{\left(a^{\prime}, s^{\prime}\right)} \mathbb{E}_{p^{\pi}}\left[\sum_{t \geq 1}^{\infty} \gamma^{t-1} r\left(s_{t}, a_{t}\right) | s_{0}=s, a_{0}=a, s_{1}=s^{\prime}, a_{1}=a^{\prime}\right] \mathbb{P}\left(s_{1}=s^{\prime}, a_{1}=a^{\prime} | s_{0}=s, a_{0}=a\right) \\ &=\sum_{\left(a^{\prime}, s^{\prime}\right)} \mathbb{E}_{p^{\pi}}\left[\sum_{t \geq 1}^{\infty} \gamma^{t-1} r\left(s_{t}, a_{t}\right) | s_{1}=s^{\prime}, a_{1}=a^{\prime}\right]  \mathbb{P}\left(s_{1}=s^{\prime}, a_{1}=a^{\prime} | s_{0}=s, a_{0}=a\right) \qquad \text{(Markov property)} \\ &=\sum_{\left(a^{\prime}, s^{\prime}\right)} \mathbb{P}\left(s_{1}=s^{\prime}, a_{1}=a^{\prime} | s_{0}=s, a_{0}=a\right) \times Q^{\pi}\left(s^{\prime}, a^{\prime}\right) \\ &=\mathbb{E}_{\left(s^{\prime}, a^{\prime}\right) \sim p^{\pi}(\cdot,\cdot | s, a)}\left[Q^{\pi}\left(s^{\prime}, a^{\prime}\right)\right]
\end{aligned}
\end{equation*}

Hence, 
<font color=blue>
$$ Q^{\pi}(s, a) = \mathbb{E}_{\left(s^{\prime}, a^{\prime}\right) \sim p^{\pi}(\cdot,\cdot | s, a)}\left[r(s, a) + \gamma Q^{\pi}\left(s^{\prime}, a^{\prime}\right)\right]$$
</font>

---------------------------------------------------------------------------------------------------------

Then, show that for the optimal policy $\pi^*$ (we assume its existence), the following holds: 

\begin{equation*}
Q^{*}(s,a)=E_{s'\sim \pi^*(.|s,a)}[r(s,a)+\gamma\max_{a'}Q^{*}(s',a')].
\end{equation*}

The correct statement should be, 
<font color=red> \begin{equation*}
Q^{*}(s,a)=E_{s'\sim p(.|s,a)}[r(s,a)+\gamma\max_{a'}Q^{*}(s',a')].
\end{equation*}
</font>

By optimality of $\pi^*$, $\pi^*$ is a deterministic map such that, $Q^{\pi^{*}}(s, \pi^{*}(s))=\max _{a^{\prime}} Q^{\pi^{*}}\left(s, a^{\prime}\right)$. Therefore, $p^{\pi^*}(s^{\prime}, a^{\prime} | s, a) = \pi^*(a^{\prime} | s^{\prime})p(s^{\prime} | s, a) = p(s^{\prime} | s, a)$ if $a^{\prime} = \pi^{*}(s^{\prime})$, $0$ otherwise. With this manipulation, we can rewrite the expectation:

\begin{equation*}
\begin{aligned} Q^{*}(s, a) &=Q^{\pi^{*}}(s, a)\\ &= \mathbb{E}_{\left(s^{\prime}, a^{\prime}\right) \sim p^{\pi^*}(\cdot,\cdot | s, a)}\left[r(s, a) + \gamma Q^{\pi^*}\left(s^{\prime}, a^{\prime}\right)\right]  \\ &= E_{s^{\prime} \sim p(\cdot | s, a)}\left[r(s, a)+\gamma Q^{\pi^{*}}\left(s^{\prime}, \pi^{*}\left(s^{\prime}\right)\right)\right] \qquad \text{(inject manipulation above)} \\ &=E_{s^{\prime} \sim p(\cdot | s, a)}\left[r(s, a)+\gamma \max _{a^{\prime}} Q^{\pi^{*}}\left(s^{\prime}, a^{\prime}\right)\right] \\ &=E_{s^{\prime} \sim p(\cdot | s, a)}\left[r(s, a)+\gamma \max _{a^{\prime}} Q^{*}\left(s^{\prime}, a^{\prime}\right)\right]
\end{aligned}
\end{equation*}

---------------------------------------------------------------------------------------------------------
---------------------------------------------------------------------------------------------------------

Finally, deduce that a plausible objective is:

\begin{equation*}
\mathcal{L}(\theta)=E_{s' \sim \pi^*(.|s,a)}\Vert r+\gamma\max\max_{a'}Q(s',a',\theta)-Q(s,a,\theta)\Vert^{2}.
\end{equation*}

The correct statement should be, 
<font color=red> \begin{equation*}
\mathcal{L}(\theta)=E_{s' \sim p(.|s,a)}\Vert r(s,a)+\gamma \max_{a'}Q(s',a',\theta)-Q(s,a,\theta)\Vert^{2}.
\end{equation*}
</font>

$Q$ is parametrized by $\theta$ and we're looking to minimize the squared error loss $||Q^*(s,a)-Q(s,a,\theta)||^2$. However $Q^*(s,a)$ is unknown, so we replace it with the following proxy $y=r(s, a)+\gamma \max_{a^{\prime}} Q\left(s^{\prime}, a^{\prime},\theta\right)$. A sensible loss is therefore, $$\mathcal L(\theta) = E_{s^{\prime} \sim p( . | s, a)}||y_t-Q(s,a,\theta)||^2$$

The DQN-learning algorithm relies on these derivations to train the parameters $\theta$ of a Deep Neural Network:

1. At the state $s_t$, select the action $a_t$ with best reward using $Q_t$ and store the results;

2. Obtain the new state $s_{t+1}$ from the environment $p$;

3. Store $(s_t,a_t,s_{t+1})$;

4. Obtain $Q_{t+1}$ by minimizing  $\mathcal{L}$ from a recovered batch from the previously stored results.

__Question 6__ Implement the class ```Memory``` that stores moves (in a replay buffer) via ```remember``` and provides a ```random_access``` to these. Specify a maximum memory size to avoid side effects. You can for example use a ```list()``` and set by default ```max_memory=100```.

````python
class Memory(object):
    """Memory Class
    Args:
                max_memory (int): Max size of this memory
    """
    
    def __init__(self, max_memory=100):
        self.max_memory = max_memory
        self.memory = deque(maxlen=max_memory)

    def remember(self, m):
        '''Store transition'''
        self.memory.append(m)

    def random_access(self, size=1):
        '''Sample a batch from the memory'''
        return random.sample(self.memory, size)
 
    def __len__(self):
        """overload length operator"""
        return(len(self.memory))
````

The pipeline we will use for training is given below:

````python
def train(agent,env,epoch,prefix=''):
    # Number of won games
    score = 0
    loss = 0
    
    # ADD
    score_per_epoch = []
    # ADD

    for e in range(epoch):
        # At each epoch, we restart to a fresh game and get the initial state
        state = env.reset()
        # This assumes that the games will terminate
        game_over = False

        win = 0
        lose = 0

        while not game_over:
            # The agent performs an action
            action = agent.act(state)

            # Apply an action to the environment, get the next state, the reward
            # and if the games end
            prev_state = state
            state, reward, game_over = env.act(action)

            # Update the counters
            if reward > 0:
                win = win + reward
            if reward < 0:
                lose = lose -reward

            # Apply the reinforcement strategy
            loss = agent.reinforce(prev_state, state,  action, reward, game_over)

        # Save as a mp4
        if e%50 == 0:
            env.draw(prefix+str(e))

        # Update stats
        score += win-lose
        
        # ADD
        score_per_epoch.append(win-lose)
        # ADD

        print("Epoch {:03d}/{:03d} | Loss {:.4f} | Win/lose count {}/{} ({})"
              .format(e, epoch, loss, win, lose, win-lose))
        agent.save(name_weights=prefix+'model.h5',name_model=prefix+'model.json')
        
    return score_per_epoch
````

__Question 7__ Implement the DQN training algorithm using a cascade of fully connected layers. You can use different learning rate, batch size or memory size parameters. In particular, the loss might oscillate while the player will start to win the games. You have to find a good criterium.

````python
class DQN(Agent):
    def __init__(self, grid_size,  epsilon = 0.1, memory_size=100, batch_size = 16,n_state=2):
        super(DQN, self).__init__(epsilon = epsilon)

        # Discount for Q learning
        self.discount = 0.99
        
        self.grid_size = grid_size
        
        # number of state
        self.n_state = n_state

        # Memory
        self.memory = Memory(memory_size)
        
        # Batch size when learning
        self.batch_size = batch_size

    def learned_act(self, s):
        return np.argmax(self.model.predict(np.expand_dims(s, 0)))
    
    def reinforce(self, s_, n_s_, a_, r_, game_over_):
        
        # first memorize the states
        self.memory.remember([s_, n_s_, a_, r_, game_over_])
        
        # second learn from the pool
        input_states = np.zeros((self.batch_size, 5,5,self.n_state))
        target_q = np.zeros((self.batch_size, self.n_action))
        
        if len(self.memory) < self.batch_size:
            return
        
        for i in range(self.batch_size):
            s_, n_s, a_, r_, game_over_ = self.memory.random_access()[0]
            input_states[i] = s_
            
            target_q[i] = self.model.predict(np.expand_dims(s_, 0))
            
            if game_over_:
                target_q[i, a_] = r_
            else:
                target_q[i, a_] = r_ + self.discount * np.max(self.model.predict(np.expand_dims(n_s, 0)))
                       
        # HINT: Clip the target to avoid exploiding gradients
        target_q = np.clip(target_q, -3, 3)
        l = self.model.train_on_batch(input_states, target_q)

        return l
    
    def save(self,name_weights='model.h5',name_model='model.json'):
        self.model.save_weights(name_weights, overwrite=True)
        with open(name_model, "w") as outfile:
            json.dump(self.model.to_json(), outfile)
            
    def load(self,name_weights='model.h5',name_model='model.json'):
        with open(name_model, "r") as jfile:
            model = model_from_json(json.load(jfile))
        model.load_weights(name_weights)
        model.compile("sgd", "mse")
        self.model = model

 
class DQN_FC(DQN):
    def __init__(self, *args, lr=0.1,**kwargs):
        super(DQN_FC, self).__init__( *args,**kwargs)
        
        model = Sequential()
        model.add(Flatten())
        model.add(Dense(64, input_shape=(5*5*self.n_state,), activation="relu"))
        model.add(Dense(16, activation="relu"))
        model.add(Dense(self.n_action, activation="linear")) 
        
        model.compile(sgd(lr=lr, decay=1e-4, momentum=0.0), "mse")
        self.model = model
````

````python
env = Environment(grid_size=size, max_time=T, temperature=0.3)
agent = DQN_FC(size, lr=.1, epsilon = 0.1, memory_size=2000, batch_size = 32)
score_per_epochs = train(agent, env, 250, prefix='experiences/DQNfc/fc_train')
````

Let us display the results. We plot the evolution of the score (with a bit of exponential smoothing) as well as a moving average of the score.

````python
def exp_smoothing(l,alpha=0.1):
    '''Exponential smoothing'''
    new_l = [l[0]]
    for i in range(1,len(l)):
        new_l.append(alpha*l[i] + (1-alpha)*new_l[-1])
    return new_l

def moving_average(l,memory):
    '''Moving average smoothing'''
    #return np.convolve(l, np.ones((memory,))/memory, mode='valid')
    cumsum = np.cumsum(np.insert(l, 0, 0)) 
    return (cumsum[memory:] - cumsum[:-memory]) / float(memory)

def plot(score, memory=1, alpha=1):
    '''Helper function to plot scores'''
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,7))
    
    ax1.plot(exp_smoothing(score,alpha))
    ax1.set_title("Score per Epoch with exponential smoothing (alpha = {})".format(alpha))
    ax1.set_xlabel("epoch")
    ax1.set_ylabel("score")

    ax2.plot(moving_average(score,memory))
    ax2.set_title("Moving average of the score (memory={})".format(memory))
    ax2.set_xlabel("epoch")
    ax2.set_ylabel("score")
    plt.show()
````

***
__Question 10__ Design a new ```train_explore``` function and environment class ```EnvironmentExploring``` to tackle the issue of exploration.

<font color=green> Remarks. </font> In addition to the improved state representation and the epsilon decay scheduling we include other improvements to help the agent reach perfect scores: we replace the DQN objective by a DDQN objective with a freezed network and we replace the sgd optimizer by RMSProp.

````python
def train_explore(agent,env,epoch,epsilon = 0.9, prefix=''):
    # Number of won games
    score = 0
    loss = 0 
    score_per_epoch = []
    agent.set_epsilon(epsilon)

    for e in range(epoch):
        state = env.reset()
        game_over = False
        win = 0
        lose = 0

        while not game_over:
            action = agent.act(state)
            prev_state = state
            state, reward, game_over = env.act(action, train=True)
            if reward > 0:
                win = win + reward
            if reward < 0:
                lose = lose - reward
            
            loss = agent.reinforce(prev_state, state,  action, reward, game_over)
            
        agent.set_epsilon(agent.epsilon*0.97)

        if e%10 == 0:
            env.draw(prefix+str(e))
        score += win-lose
        score_per_epoch.append(win-lose)
        print("Epoch {:03d}/{:03d} | Loss {:.4f} | Win/lose count {}/{} ({})"
              .format(e, epoch, loss, win, lose, win-lose))
        agent.save(name_weights=prefix+'model.h5',name_model=prefix+'model.json')
        
    return score_per_epoch
````

````python
class EnvironmentExploring(object):
    def __init__(self, grid_size=10, max_time=500, temperature=0.1):
        
        grid_size = grid_size+4
        self.grid_size = grid_size
        self.max_time = max_time
        self.temperature = temperature
        self.board = np.zeros((grid_size,grid_size))
        self.position = np.zeros((grid_size,grid_size))
        
        ## new state
        self.malus_position = np.zeros((grid_size,grid_size))
        ## new state

        self.x = 0
        self.y = 1
        self.t = 0
        self.scale=16
        self.to_draw = np.zeros((max_time+2, grid_size*self.scale, grid_size*self.scale, 3))
        
    def draw(self,e):
        skvideo.io.vwrite(str(e) + '.mp4', self.to_draw)

    def get_frame(self,t):
        '''Called before every taken action to store the video matrix'''
        b = np.zeros((self.grid_size,self.grid_size,3))+128
        b[self.board>0,0] = 256       
        b[self.board < 0, 2] = 256    
        b[self.x,self.y,:]=256        
        b[-2:,:,:]=0                  
        b[:,-2:,:]=0
        b[:2,:,:]=0
        b[:,:2,:]=0
        b =  cv2.resize(b, None, fx=self.scale, fy=self.scale, interpolation=cv2.INTER_NEAREST)
        self.to_draw[t,:,:,:]= b 


    def act(self, action, train=False):
        """This function returns the new state, reward and decides if the
        game ends.
        Actions:
                        N.1
                    O.3     E.2
                        S.0
        """
        self.get_frame(int(self.t)) 
        self.position = np.zeros((self.grid_size, self.grid_size))
        self.position[0:2,:]= -1
        self.position[:,0:2] = -1
        self.position[-2:, :] = -1
        self.position[:, -2:] = -1
        self.position[self.x, self.y] = 1
        if action == 0: 
            if self.x == self.grid_size-3:
                self.x = self.x-1
            else:
                self.x = self.x + 1
        elif action == 1:
            if self.x == 2:
                self.x = self.x+1
            else:
                self.x = self.x-1
        elif action == 2: 
            if self.y == self.grid_size - 3:
                self.y = self.y - 1
            else:
                self.y = self.y + 1
        elif action == 3:
            if self.y == 2:
                self.y = self.y + 1
            else:
                self.y = self.y - 1
        else:
            RuntimeError('Error: action not recognized')

        self.t = self.t + 1
        reward = 0
        
        ## new state
        if train:
            reward = -self.malus_position[self.x, self.y]
        self.malus_position[self.x, self.y] += 0.1
        ## new state
        
        reward = reward + self.board[self.x, self.y]
        self.board[self.x, self.y] = 0    
        game_over = self.t > self.max_time 
        
        ## new state
        state = np.concatenate((self.malus_position.reshape(self.grid_size, self.grid_size,1),
                                        self.board.reshape(self.grid_size, self.grid_size,1),
                                self.position.reshape(self.grid_size, self.grid_size,1)),axis=2)
        ## new state
        
        state = state[self.x-2:self.x+3,self.y-2:self.y+3,:]

        return state, reward, game_over
    
    def reset(self):
        """This function resets the game and returns the initial state"""
        self.x = np.random.randint(3, self.grid_size-3, size=1)[0]
        self.y = np.random.randint(3, self.grid_size-3, size=1)[0]
        bonus = 0.5*np.random.binomial(1,self.temperature,size=self.grid_size**2)
        bonus = bonus.reshape(self.grid_size,self.grid_size)
        malus = -1.0*np.random.binomial(1,self.temperature,size=self.grid_size**2)
        malus = malus.reshape(self.grid_size, self.grid_size)
        self.to_draw = np.zeros((self.max_time+2, self.grid_size*self.scale, self.grid_size*self.scale, 3))
        malus[bonus>0]=0
        self.board = bonus + malus
        self.position = np.zeros((self.grid_size, self.grid_size))
        self.position[0:2,:]= -1 
        self.position[:,0:2] = -1
        self.position[-2:, :] = -1
        self.position[:, -2:] = -1 
        self.board[self.x,self.y] = 0 
        self.t = 0
        #self.position[self.x,self.y] = 1
        self.malus_position = np.zeros((self.grid_size, self.grid_size))

        # new state
        state = np.concatenate((self.board.reshape(self.grid_size, self.grid_size,1),
                                self.position.reshape(self.grid_size, self.grid_size,1),
                                self.malus_position.reshape(self.grid_size, self.grid_size,1)),axis=2)
        # new state
        
        state = state[self.x - 2:self.x + 3, self.y - 2:self.y + 3, :]
                
        return state
````

````python
# Training
epoch = 200
env = EnvironmentExploring(grid_size=size, max_time=T, temperature=0.3)
agent = DQN_CNN(size, lr=.1, epsilon = 0.9, memory_size=50000, batch_size = 32,n_state=3)
score = train_explore(agent, env, epoch, prefix='experiences/DQNcnn_explore4/cnn_train_explore')
````
